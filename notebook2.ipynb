```python
import pandas as pd
import warnings

import matplotlib.pyplot as plt
from sklearn import tree
import seaborn as sns
import sklearn
import numpy as np

wines = pd.read_csv("csv_dateien/winequality-red.csv")

# fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(15, 5))

sns.heatmap(wines[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol","quality"]].corr(), annot=True, center=0)

# wines[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol","quality"]]

g = sns.jointplot(x="alcohol", y="quality", data=wines[["alcohol", "quality"]],
                  kind="reg", truncate=False,
                  xlim=(8.4, 14.9), ylim=(0, 10),
                  color="m", height=7)

g2 = sns.jointplot(x="volatile acidity", y="quality", data=wines[["volatile acidity", "quality"]],
                  kind="reg", truncate=False,
                  xlim=(0, 2), ylim=(0, 10),
                  color="m", height=7)

g3 = sns.jointplot(x="fixed acidity", y="pH", data=wines[["fixed acidity", "pH"]],
                  kind="reg", truncate=False,
                  xlim=(4, 16), ylim=(2, 5),
                  color="m", height=7)
```

```python
import pandas as pd
import warnings

import matplotlib.pyplot as plt
from sklearn import tree
import seaborn as sns
import sklearn
import numpy as np

wines = pd.read_csv("csv_dateien/Hitters.csv")
wines["general_performance"] = wines["Hits"]+wines["HmRun"]+wines["Runs"]+wines["Walks"]+0.8*wines["PutOuts"]
# fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(15, 5))

sns.heatmap(wines[["Hits","HmRun","Runs","Walks","PutOuts","Assists","Errors","Salary","general_performance"]].corr(), annot=True, center=0)


# wines[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol","quality"]]
g = sns.jointplot(x="Salary", y="general_performance", data=wines[["Salary", "Hits", "AtBat", "general_performance"]],
                  kind="reg", truncate=False,
                  color="m", height=7)
```

```python
import pandas as pd
import warnings

import matplotlib.pyplot as plt
from sklearn import tree
import seaborn as sns
import sklearn
import numpy as np

wines = pd.read_csv("csv_dateien/Laptop_Information.csv")
# wines["general_performance"] = wines["Hits"]+wines["HmRun"]+wines["Runs"]+wines["Walks"]+0.8*wines["PutOuts"]
# fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(15, 5))

sns.heatmap(wines[["Company","Rating","No_of_ratings","Review","Size","Processor","RAM","Memory","OpSys","Price","MRP","ImgURL"]].corr(), annot=True, center=0)


# wines[["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides","free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol","quality"]]
g = sns.jointplot(x="Price", y="RAM", data=wines[["Price", "Rating", "RAM"]],
                  kind="reg", truncate=False,
                  color="m", height=7)

# Draw a nested boxplot to show bills by day and time
sns.boxplot(x="RAM", y="Price",
            palette=["m", "g"],
            data=wines)
sns.despine(offset=10, trim=True)
```
