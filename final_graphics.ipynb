```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('csv_dateien/starcraftFinalcsvs/stackedRunSup.csv')

df = df[df["fraction"] != "0"]
```

```python


custom_palette = ["#dd8452","#4c72b0",  "#55a868"]
sns.set_theme(style="ticks")
# Show the joint distribution using kernel density estimation
g = sns.jointplot(
    data=df, palette=custom_palette,
    x="total_workers", y="total_army", hue="fraction", ylim=[None, 170],
    kind="kde", height=10
)
```

```python
df = df[df["resource_mining"] < 200000]
custom_palette = ["#55a868", "#dd8452","#4c72b0"]
hue_order = ['Zerg', 'Protoss', 'Terraner']
g3= sns.lmplot(x="total_workers", y="resource_mining", hue="fraction", hue_order=hue_order, palette=custom_palette,  data=df,
          height=8)
```
