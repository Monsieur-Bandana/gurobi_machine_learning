```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('csv_dateien/starcraftFinalcsvs/stackedRunSup.csv')

df = df[df["fraction"] != "0"]
```

```python


custom_palette = ["#dd8452","#4c72b0",  "#55a868"]
sns.set_theme(style="ticks")
# Show the joint distribution using kernel density estimation
g = sns.jointplot(
    data=df, palette=custom_palette,
    x="total_workers", y="total_army", hue="fraction", ylim=[None, 170],
    kind="kde", height=10
)
```

```python
df = df[df["resource_mining"] < 200000]
custom_palette = ["#55a868", "#dd8452","#4c72b0"]
hue_order = ['Zerg', 'Protoss', 'Terraner']
g3= sns.lmplot(x="total_workers", y="resource_mining", hue="fraction", hue_order=hue_order, palette=custom_palette,  data=df,
          height=8)
```

```python
df2 = pd.read_csv('csv_dateien/starcraftFinalcsvs/allRunsAll.csv')

# df2 = df2[df2["resource_mining"] < 200000]
custom_palette = ["#55a868", "#dd8452","#4c72b0"]
hue_order = ['Zerg', 'Protoss', 'Terraner']
g3= sns.lmplot(x="total_workers", y="time", hue="fraction", hue_order=hue_order, palette=custom_palette,  data=df2,
          height=8)
```

```python
df2["ratio"] = df2["total_workers"]/df2["total_army"]
df2 = df2[df2["ratio"] < 4]
custom_palette = ["#55a868", "#dd8452","#4c72b0"]
hue_order = ['Zerg', 'Protoss', 'Terraner']
g3= sns.lmplot(x="ratio", y="time", hue="fraction", hue_order=hue_order, palette=custom_palette,  data=df2,
          height=8)
```

```python
dfProtoss = df2[df2["fraction"] == "Protoss"]
dfZerg = df2[df2["fraction"] == "Zerg"]
dfTerran = df2[df2["fraction"] == "Terraner"]

corrProtoss = dfProtoss["ratio"].corr(dfProtoss["time"])
corrZerg = dfZerg["ratio"].corr(dfZerg["time"])
corrTerran = dfTerran["ratio"].corr(dfTerran["time"])

print("corrProtoss: {}\ncorrZerg: {}\ncorrTerran: {}".format(corrProtoss,corrZerg,corrTerran))
```
